1. Find The Occurrence Of the Employee Names (Top 5)

```python
top_5_employees = df['EmployeeName'].value_counts().head(5)
print(top_5_employees)
```

2. Find The Number Of Unique Job Titles

```python
unique_job_titles = df['JobTitle'].nunique()
print(f'Number of unique job titles: {unique_job_titles}')
```

3.  Find The Total Number Of Job Titles Containing "captain" 

```python
captain_count = df['JobTitle'].str.contains('captain', case=False).sum()
print(f'Total number of job titles containing "captain": {captain_count}')
```

4. Display All the Employee Names from Fire Department

```python
fire_department_employees = df[df['Agency'].str.contains('Fire Department', case=False)]['EmployeeName']
print(fire_department_employees)
```

5. Find The Minimum, Maximum, and Average BasePay

```python
min_basepay = df['BasePay'].min()
max_basepay = df['BasePay'].max()
avg_basepay = df['BasePay'].mean()
print(f'Min BasePay: {min_basepay}, Max BasePay: {max_basepay}, Avg BasePay: {avg_basepay}')
```

6. Replace 'Not provided' in EmployeeName Column with NaN

```python


df['EmployeeName'].replace('Not provided', np.nan, inplace=True)
```

7. Drop The Rows Having 5 Missing Values

```python
df.dropna(thresh=len(df.columns) - 5, inplace=True)
```

8. Find The Job Title Of ALBERT PARDINI

```python
albert_job_title = df[df['EmployeeName'] == 'ALBERT PARDINI']['JobTitle']
print(albert_job_title)
```

```python
9. How Much ALBERT PARDINI Makes (Include Benefits)
```

```python
albert_salary = df[df['EmployeeName'] == 'ALBERT PARDINI']['TotalPayBenefits']
print(f'ALBERT PARDINI makes: {albert_salary.values[0]}')
```

10. Display Name Of The Person Having The Highest Basepay

```python
highest_basepay_employee = df.loc[df['BasePay'].idxmax()]['EmployeeName']
print(f'The person with the highest BasePay is: {highest_basepay_employee}')
```

11. Find The Average Basepay of All Employees Per Year

```python
avg_basepay_per_year = df.groupby('Year')['BasePay'].mean()
print(avg_basepay_per_year)
```

12. Find Average Basepay Of All Employees Per Job Title

```python
avg_basepay_per_job_title = df.groupby('JobTitle')['BasePay'].mean()
print(avg_basepay_per_job_title)
```

13. Find The Average Basepay of All Employees Having Job Title Accountant

```python
avg_basepay_accountant = df[df['JobTitle'] == 'Accountant']['BasePay'].mean()
print(f'Average BasePay for Accountants: {avg_basepay_accountant}')
```
